 "While directly displaying images using some OpenCV functions might have different behavior in Colab, libraries like Matplotlib can be used for visualizing image data processed with OpenCV."



```python
from matplotlib import pyplot as plt
```

When you read an image using OpenCV's cv2.imread() function, it loads the image data in the BGR color format (Blue, Green, Red). However, libraries like Matplotlib, which are commonly used for displaying images in Python, expect the image data in the RGB color format (Red, Green, Blue).

To correctly display an image read by OpenCV using Matplotlib, you need to convert the color format from BGR to RGB. This is because the order of color channels is different between the two formats. If you don't perform this conversion, the colors in the displayed image will appear inverted (e.g., red will appear blue, and blue will appear red).

The conversion involves swapping the Red and Blue color channels. OpenCV provides functions like cv2.cvtColor() that can perform this conversion easily. You would typically use cv2.cvtColor(img, cv2.COLOR_BGR2RGB) where img is your image data in BGR format, and cv2.COLOR_BGR2RGB is a flag that specifies the conversion from BGR to RGB.

```python
img=cv2.imread("/content/my post.jpg")

# convert BGR to RGB
img_rgb= cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

# Convert to Grayscale
img_gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
# Apply Canny edge detection
low_threshold=100
high_threshold=200
img_canny=cv2.Canny(img_gray, low_threshold, high_threshold)

# Print Shapes
print("Orignal Image Shape:", img.shape)
print("GrayScale Image Shape:", img_gray.shape)
print("Canny Image Shape:", img_canny.shape)


# plot the image side by side
plt.figure(figsize=(10,5))

# Display RGB image
plt.subplot(1,2,1)
plt.imshow(img_rgb)
plt.title("Original RGB Image")
plt.axis('off')

# Display GrayScale image
plt.subplot(1,3,3)
plt.imshow(img_canny, cmap='gray')  # <-- Use cmap='gray'
# plt.imshow(img_gray)
plt.title("Canny Edge Detection")
plt.axis('off')
plt.show()
```
