```python
plt.subplot(1,2,1)
plt.imshow(img)
plt.title("Original Image")
plt.axis('off')
plt.show()

plt.subplot(1,2,2)
plt.imshow(img_rgb)
plt.title("Original RGB Image")
plt.axis('off')
plt.show()
```

```python
# Define source and destination points
pts1= np.float32([[2000,118],[1120,265],[540,668],[871,830]])
pts2= np.float32([[0,0],[width,0],[0,height],[width,height]])
```

```python
# Get perspective transform matrix and apply wrap
matrix=cv2.getPerspectiveTransform(pts1,pts2)
img_out=cv2.warpPerspective(img, matrix, (width, height))
```

```python
img_out_rgb= cv2.cvtColor(img_out,cv2.COLOR_BGR2RGB)

# display original an warped images
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.imshow(img_rgb)
plt.title("Original Image")
plt.axis('off')
plt.show()

plt.subplot(1,2,2)
plt.imshow(img_out_rgb)
plt.title("Original Image")
plt.axis('off')
plt.show()


# wil continue later it is not working
```
